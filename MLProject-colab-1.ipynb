```python
# Merging Bureau data with Bureau_balance data using "SK_BUREAU" coloumn as the reference 
bur_joined = pd.merge(bur, bur_bal, on ='SK_ID_BUREAU', how ='inner')
```

```python
# Taking only subset of columns from the combined Bureau data 
bur_joined_1 = bur_joined[['SK_ID_CURR','AMT_CREDIT_MAX_OVERDUE', 'CNT_CREDIT_PROLONG','AMT_CREDIT_SUM', 'AMT_CREDIT_SUM_DEBT', 'AMT_CREDIT_SUM_OVERDUE','MONTHS_BALANCE' ]]
```

```python
# Using Groupby to find the Sum of values per id (SK_ID_CURR)
bur_joined_agg = bur_joined_1.groupby('SK_ID_CURR').sum().reset_index()
```
